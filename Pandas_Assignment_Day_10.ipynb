## Groupby

```python
# Create a dataframe using dictionary of your choice

# data =

data = pd.DataFrame({
    'Category': ['A', 'B', 'A', 'B', 'A', 'C'],
    'Value': [10, 15, 12, 18, 11, 20],
    'Quantity': [1, 2, 1, 3, 2, 1]
})
```

```python
# Use Groupby of single column with aggregate sum()

print(data.groupby("Category")[["Value"]].sum())
```

```python
# Use Groupby of single column with aggregate count()

print(data.groupby("Category")[["Value"]].count())
```

```python
# Use Groupby of single column with aggregate min() and max()
print(data.groupby("Category")[["Value"]].min())
print(data.groupby("Category")[["Value"]].max())
```

```python
# Use Groupby of any 2 columns with aggregate mean()
print(data.groupby(["Category","Value"])[["Quantity"]].mean())
```

```python
# Use Groupby of any 2 columns with aggregate min() and max()

print(data.groupby(["Category","Value"])[["Quantity"]].min())
```
